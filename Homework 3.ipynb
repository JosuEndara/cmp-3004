- Given an array of integers, calculate the average

````python
.data
array:  .word 10, 20, 30, 40, 50    # Ejemplo de array de enteros
size:   .word 5                      # tamanio del array
result: .asciiz "Average: "      #Para imprimir el promedio

.text
.globl main

main:
    # Calcula la suma de los numeros
    li $t0, 0                   # Inicializar suma en 0
    la $t1, array               # Carga direccion base del arreglo
    lw $t2, size                # Carga el tamanio del arreglo
    li $t3, 0                   # Contador del loop

sum_loop:
    beq $t3, $t2, calculate_average   # Salto del loop cuando contador sea igual a tamanio
    lw $t4, 0($t1)                    # Carga valor actual de iteracion del arreglo
    add $t0, $t0, $t4                 # Suma el valor a la suma total
    addi $t1, $t1, 4                  # Incremento puntero del arreglo
    addi $t3, $t3, 1                  # Incremento puntero del contador
    j sum_loop			# SALTO continuacion loop

calculate_average:
    # Calcula el promedio
    div $t0, $t2                 # Divide suma para el tamanio
    mflo $t5                     # Guarda el resultado

    # Muestra el texto de Average:
    li $v0, 4
    la $a0, result
    syscall

    # Muestra el valor del promedio
    li $v0, 1
    move $a0, $t5
    syscall

    #SALIDA PROGRAMA
    li $v0, 10
    syscall
````

- Write a program to transform from Celsius to Fahrenheit and vice versa

````python
.data
prompt:     .asciiz "Enter the temperature value: " #Auxiliares de textos para imprimir resultados y pedir input 
result:     .asciiz "Result: "
newline:    .asciiz "\n"

.text
.globl main

main:
    # Impresion que pide al usuario un valor de temperatura
    li $v0, 4
    la $a0, prompt
    syscall

    # Lectura de input de usuario
    li $v0, 5
    syscall
    move $t0, $v0  # Guarda el valor de input

    # Convertir de Celsius a Fahrenheit
    li $v0, 4
    la $a0, result
    syscall

    mul $t1, $t0, 9   # Multiplicar Celsius temperatura por 9
    div $t1, $t1, 5   # Dividir este resultado para 5
    addi $t1, $t1, 32 # Sumar 32 para obtener la temperatura en Fahrenheit

    move $a0, $t1     # Se mueve la temperatura Fahrenheit a $a0, para impresion
    li $v0, 1
    syscall
    
    li $v0, 4
    la $a0, newline
    syscall

    # Convertir de Fahrenheit a Celsius
    li $v0, 4
    la $a0, result
    syscall

    subi $t2, $t1, 32 # Restar 32 de la temperatura Fahrenheit
    mul $t2, $t2, 5   # Multiplicar el resultado por 5
    div $t2, $t2, 9   # Dividir para 9

    move $a0, $t2     # Se mueve la temperatura en Celcius a $a0
    li $v0, 1
    syscall


    li $v0, 4
    la $a0, newline
    syscall

    # SALIDA PROGRAMA
    li $v0, 10
    syscall
````

### 4. What are the differences among sequential access, direct access, and random access?

ACCESO SECUENCIAL: En este tipo de acceso se accede a los datos como si fuera una secuencia lineal, esto quiere decir que los datos tienen un orden de acceso.

ACCESO DIRECTO: En este tipo de acceso se hace uso la ubicacion fisica unica de almacenamiento para acceder a los registros en especifico, para lograr esto se hace uso de la direccion especifica de memoria.

### 5. What common characteristics are shared by all RAID levels?

Los niveles de RAID comparten características comunes que contribuyen a su funcionalidad y beneficios. Estas características incluyen:

    1. Utilización de varios discos: Los niveles de RAID utilizan múltiples discos en su configuración, lo que proporciona una mayor capacidad de almacenamiento y un mejor rendimiento.

    2. Redundancia: Los niveles de RAID incorporan la redundancia para proteger los datos. Esto se logra almacenando información de paridad o duplicando los datos en diferentes discos, lo que permite recuperar la información en caso de fallo de una unidad.

    3. Data striping: Esta característica implica dividir los datos en bloques y distribuirlos en varios discos. Esto mejora las velocidades de lectura y escritura al permitir el acceso simultáneo a los datos desde diferentes discos.

En resumen, los niveles de RAID utilizan múltiples discos, incorporan redundancia para proteger los datos y utilizan data striping para mejorar el rendimiento. Estas características hacen que RAID sea una tecnología efectiva para aumentar la capacidad de almacenamiento, garantizar la integridad de los datos y mejorar el rendimiento del sistema.
